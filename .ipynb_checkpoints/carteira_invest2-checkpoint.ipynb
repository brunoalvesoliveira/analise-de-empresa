```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns
from matplotlib import style
```

```python
style.use('seaborn-muted')
```

Buscando dados 

```python
tickers = ['TOTS3.SA', '^BVSP']
dados = pd.DataFrame()

for ticker in tickers:
    dados[ticker] = yf.download(ticker, start='2019-01-01')['Adj Close']

dados.head()
```

Gráfico: cotação

```python
plt.figure(figsize=(12, 6));
plt.title('Cotação da TOTVS versus pontos do IBOVESPA', fontsize=15);
x, = plt.plot(dados['TOTS3.SA'], color='orange');
y, = plt.plot(dados['^BVSP'], color='blue');
plt.xlabel('Data', fontsize=13);
plt.ylabel('Cotação', fontsize=13);
plt.legend([x, y], ['Cotação TOTVS', 'Pontos IBOVESPA'], loc=2, fontsize=12);
```

Gráfico: Cotação da TOTVS versus Pontos do IBOVESPA de outra forma

```python
fig, ax1 = plt.subplots(figsize=(12, 6));

color = 'blue';
ax1.set_title('Cotação da TOTVS versus Pontos do IBOVESPA');
ax1.text(np.datetime64('2021-01-01'), 70000, 'By: Invest Jr. - Yahoo Finance', fontsize=12);
ax1.set_xlabel('Data', fontsize=13);
ax1.set_ylabel('Pontos - IBOVESPA', fontsize=13, color='black');
ax1.plot(dados['^BVSP'], color=color);
ax1.tick_params(axis='y', labelcolor=color);

ax2 = ax1.twinx()

color = 'darkorange';
ax2.set_ylabel('Cotação - TOTVS', color='black');
ax2.plot(dados['TOTS3.SA'], color=color);
ax2.tick_params(axis='y', labelcolor=color);

fig.tight_layout()
```

Calculando retorno simples

```python
retorno_simples = dados.pct_change()
retorno_simples.head()
```

Calculando retorno acumulado e normalizando a base de dados

```python
retorno_acumulado = (1 + retorno_simples).cumprod()
retorno_acumulado.iloc[0] = 1
retorno_acumulado.head()
```

Gráfico: retorno

```python
plt.figure(figsize=(12, 6));
plt.title('Retorno acumulado: TOTVS versus IBOVESPA', fontsize=15);
x, = plt.plot(retorno_acumulado['TOTS3.SA'], color='orange');
y, = plt.plot(retorno_acumulado['^BVSP'], color='blue');
plt.xlabel('Data', fontsize=13);
plt.ylabel('Retorno acumulado', fontsize=13);
plt.text(np.Datetime64('2021-01-01'), 0.8, 'By: Invest Jr. - Yahoo Finance', fontsize=10);
plt.legend([x, y], ['Retorno acumulado - TOTVS', 'Retorno acumulado - IBOVESPA'], loc=2, fontsize=12);
```

Correlação média dos retornos acumulados das variáveis 

```python
corr = retorno_acumulado.corr()
corr
```

Criando mapa de calor

```python
plt.figure(figsize=(12, 6));
sns.heatmap(corr, annot=True, cmap='coolwarm');
plt.title('Correlação entre os retornos acumulados', fontsize=15);
```
